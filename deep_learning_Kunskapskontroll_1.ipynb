```python
#It is possible to load pretrained models which is a powerful concept.
#Model implementation: https://keras.io/api/applications/#usage-examples-for-image-classification-models

#Class list for ResNet50, for example tennis ball is 852: https://deeplearning.cms.waikato.ac.nz/user-guide/class-maps/IMAGENET/3
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
# img_path = 'zebra.jpg'
# img_path = 'linjal.jpg'
img_path = 'cat.jpg'
#img_path = 'boat.jpg'
#img_path = 'cup.jpg'

img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)


```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

2. Answer the question: Why is "Pretrained models" a very powerful concept?

"Pretrained models" har fördelarna:

1. Eftersom modellen redan är tränad krävs inte stora mängder data för att träna den från grunden.

2. Genom att dra nytta av det omfattande och mångfacetterade datasetet som modellen tränades på, kan den leverera överlägsen prestanda jämfört med modeller tränade på begränsade egna dataset.

3. Användningen av en förtränad modell möjliggör snabb implementering och användning jämfört med den tidskrävande processen att träna en modell från början.
